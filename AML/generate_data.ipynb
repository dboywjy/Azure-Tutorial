```python
import sklearn
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

# 
n_samples = 1000  # 
n_features = 4  # 
n_classes = 3  # 

X, y = make_classification(
    n_samples=n_samples,
    n_features=n_features,
    n_informative=n_features,
    n_redundant=0,
    n_classes=n_classes,
    random_state=42
)

# 
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Random Iris Data')
plt.show()
```

```python
X = pd.DataFrame(X).rename(columns={col:"COL_"+str(col) for col in range(0, 4)})
y = pd.DataFrame(y).rename(columns={0:"Y"})
data = pd.concat([X, y], axis=1)
```

```python
data.to_csv("./iris_data.csv", index=False)
```
